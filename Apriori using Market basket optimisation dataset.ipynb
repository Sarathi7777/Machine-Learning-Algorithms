```python
df=df.fillna(0)
```

```python
transactions=[]
for i in range(len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
transactions[0]
```

```python
from apyori import apriori
rules=apriori(transactions,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
result=list(rules)
d=pd.DataFrame(result)
d.head()
```

```python
support=d.support
first_values=[]
second_values=[]
third_values=[]
fourth_values=[]
for i in range(d.shape[0]):
  single_list=d['ordered_statistics'][i][0]
  first_values.append(list(single_list[0]))
  second_values.append(list(single_list[1]))
  third_values.append(single_list[2])
  fourth_values.append(single_list[3])
```

```python
lhs=pd.DataFrame(first_values)
rhs=pd.DataFrame(second_values)
confidence=pd.DataFrame(third_values,columns=['confidance'])
lift=pd.DataFrame(fourth_values,columns=['lift'])
```

```python
df_final=pd.concat([lhs,rhs,support,confidence,lift],axis=1)
df_final
```

```python
def inspect(results):
  lhs=[tuple(i[2][0][0])[0] for i in result]
  rhs=[tuple(i[2][0][1])[0] for i in result]
  support=[i[1] for i in results]
  confidence=[i[2][0][2] for i in result]
  lift=[i[2][0][3] for i in result]
  return list(zip(lhs,rhs,support,confidence,lift))
d=pd.DataFrame(inspect(result),columns=['Item1','Item2','Support','confidence','lift'])
d
```
